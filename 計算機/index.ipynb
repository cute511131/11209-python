```python
import tkinter as tk
from tkinter import ttk


class CalculatorApp:
    def __init__(self, master):
        self.master = master
        self.master.title("工程用計算機")

        self.entry_var = tk.StringVar()
        self.entry = ttk.Entry(
            self.master,
            textvariable=self.entry_var,
            font=("Arial", 14),
            justify="right",
        )
        self.entry.grid(row=0, column=0, columnspan=4, sticky="nsew")

        self.create_buttons()

    def create_buttons(self):
        buttons = [
            "7",
            "8",
            "9",
            "/",
            "4",
            "5",
            "6",
            "*",
            "1",
            "2",
            "3",
            "-",
            "0",
            ".",
            "+",
        ]

        row_val = 1
        col_val = 0

        for button in buttons:
            ttk.Button(
                self.master,
                text=button,
                command=lambda button=button: self.on_click(button),
            ).grid(row=row_val, column=col_val)
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

        # 清除和計算按鈕
        ttk.Button(self.master, text="C", command=self.clear_entry).grid(
            row=row_val, column=col_val
        )
        ttk.Button(self.master, text="=", command=self.calculate).grid(
            row=row_val, column=col_val + 1
        )

    def on_click(self, button_value):
        current = self.entry_var.get()
        new_text = current + str(button_value)
        self.entry_var.set(new_text)

    def clear_entry(self):
        self.entry_var.set("")

    def calculate(self):
        try:
            result = eval(self.entry_var.get())
            self.entry_var.set(result)
        except Exception as e:
            self.entry_var.set("錯誤")


if __name__ == "__main__":
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from sympy import *


class CalculatorApp:
    def __init__(self, master):
        self.master = master
        self.master.title("微積分計算機")

        self.entry_var = tk.StringVar()
        self.entry = ttk.Entry(
            self.master,
            textvariable=self.entry_var,
            font=("Arial", 14),
            justify="right",
        )
        self.entry.grid(row=0, column=0, columnspan=4, sticky="nsew")

        self.create_buttons()

    def create_buttons(self):
        buttons = [
            "7",
            "8",
            "9",
            "/",
            "4",
            "5",
            "6",
            "*",
            "1",
            "2",
            "3",
            "-",
            "0",
            ".",
            "+",
            "x",
            "^",
            "sqrt",
            "(",
            ")",
        ]

        row_val = 1
        col_val = 0

        for button in buttons:
            ttk.Button(
                self.master,
                text=button,
                command=lambda button=button: self.on_click(button),
            ).grid(row=row_val, column=col_val)
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

        # 清除和計算按鈕
        ttk.Button(self.master, text="C", command=self.clear_entry).grid(
            row=row_val, column=col_val
        )
        ttk.Button(self.master, text="=", command=self.calculate).grid(
            row=row_val, column=col_val + 1
        )

    def on_click(self, button_value):
        current = self.entry_var.get()

        if button_value == "x":
            button_value = "*"

        if button_value == "^":
            button_value = "**"

        if button_value == "sqrt":
            button_value = "sqrt("

        new_text = current + str(button_value)
        self.entry_var.set(new_text)

    def clear_entry(self):
        self.entry_var.set("")

    def calculate(self):
        try:
            expression = self.entry_var.get()
            result = eval(expression)
            self.entry_var.set(result)
        except Exception as e:
            self.entry_var.set("錯誤")


if __name__ == "__main__":
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()
```
